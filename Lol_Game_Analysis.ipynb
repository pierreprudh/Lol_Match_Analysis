<h3 style="color:#c38d9e; font-family: Arial, sans-serif; font-size: 24px; margin-top: 40px; text-align: center;">
print_match_history
</h3>

```python
def print_match_history(match_detail) :
    team_a_summoner_names = []
    team_a_champion_names = []
    team_b_summoner_names = []
    team_b_champion_names = []

    for participant in match_detail['info']['participants']:
        if participant['teamId'] == 100:
            team_a_summoner_names.append(participant['summonerName'])
            team_a_champion_names.append(participant['championName'])
        else:
            team_b_summoner_names.append(participant['summonerName'])
            team_b_champion_names.append(participant['championName'])

    print('Team A:')
    for i in range(len(team_a_summoner_names)):
        print(f"Summoner: {team_a_summoner_names[i]} --> Champion: {team_a_champion_names[i]}")

    print('Team B:')
    for i in range(len(team_b_summoner_names)):
        print(f"Summoner: {team_b_summoner_names[i]} --> Champion: {team_b_champion_names[i]}")
```

<h2 style="color:#f7ce5b; font-family: Arial, sans-serif; font-size: 40px; margin-top: 50px; border-bottom: 2px solid #f7ce5b;">
Image Printing
</h2>

<h3 style="color:#c38d9e; font-family: Arial, sans-serif; font-size: 24px; margin-top: 40px; text-align: center;">
print_match_history_img
</h3>

```python
from PIL import Image, ImageDraw, ImageFont

def print_match_history_img(match_detail, output_path):
    team_a_summoner_names = []
    team_a_champion_names = []
    team_b_summoner_names = []
    team_b_champion_names = []

    roles = ['Top', 'Jungle', 'Mid', 'ADC', 'Support']

    for participant in match_detail['info']['participants']:
        if participant['teamId'] == 100:
            team_a_summoner_names.append(participant['summonerName'])
            team_a_champion_names.append(participant['championName'])
        else:
            team_b_summoner_names.append(participant['summonerName'])
            team_b_champion_names.append(participant['championName'])

    img = Image.new('RGB', (800, 500), color='white')
    draw = ImageDraw.Draw(img)

    font = ImageFont.truetype('Helvetica', 24)
    y = 50
    draw.text((50, y), 'Team A:', font=font, fill='black')
    y += 30
    for i in range(len(team_a_summoner_names)):
        draw.text((50, y), f"{roles[i]}: {team_a_summoner_names[i]} --> Champion: {team_a_champion_names[i]}", font=font, fill='black')
        y += 30

    y += 30
    draw.text((50, y), 'Team B:', font=font, fill='black')
    y += 30
    for i in range(len(team_b_summoner_names)):
        draw.text((50, y), f"{roles[i]}: {team_b_summoner_names[i]} --> Champion: {team_b_champion_names[i]}", font=font, fill='black')
        y += 30

    img.save(output_path)
```

```python
for i in range(0, 5):
    match_detail = watcher.match.by_id(region, match_history[i])
    output_path = f"./results/match_{i+1}.png"
    print_match_history_img(match_detail, output_path)

Image.open('./results/match_1.png')
```

</h2>

<h3 style="color:#c38d9e; font-family: Arial, sans-serif; font-size: 24px; margin-top: 40px; text-align: center;">
player_analysis
</h3>

```python
from collections import Counter
import matplotlib.pyplot as plt
from PIL import Image
import requests
from io import BytesIO
from matplotlib.offsetbox import BboxImage
import numpy as np

match_history = watcher.match.matchlist_by_puuid('euw1', summoner['puuid'])

def player_analysis(match_history, name) :

    summoners_name = []

    for i in range(len(match_history)) :
        match_details = watcher.match.by_id(region, match_history[i])
        participants = match_details['info']['participants']
        for participant in participants:
            summoners_name.append(participant['summonerName'])
        
    while name in summoners_name:
        summoners_name.remove(name)

    summoners_count = Counter(summoners_name)
    top_names = summoners_count.most_common(5)

    names_top = [name[0] for name in top_names]
    counts_top = [name[1] for name in top_names]

    fig, ax = plt.subplots()
    ax.barh(names_top, counts_top)

    for i, count in enumerate(counts_top):
        ax.text(count + 1, i, str(count), ha='left', va='center')

    images = []

    ax.set_xlabel('Nombre d\'apparitions')
    ax.set_ylabel('Nom de l\'invocateur')
    ax.set_title('Les 5 invocateurs les plus courants')
    ax.invert_yaxis()
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    versions = watcher.data_dragon.versions_for_region(region)
    latest_version = versions['n']['profileicon']

    for i, name in enumerate(names_top):
        participant = watcher.summoner.by_name(region, name)
        if 'profileIconId' in participant:
            icon_id = participant['profileIconId']
            icon_url = f'http://ddragon.leagueoflegends.com/cdn/{latest_version}/img/profileicon/{icon_id}.png'
            response = requests.get(icon_url)
            img = Image.open(BytesIO(response.content))
            images.append(img)
        else:
            print(f"No profile icon found for {name}")
            images.append(None)

    for name, count, image in zip(names, counts, images):
        img = image.resize((30, 30), resample=Image.BICUBIC)

        img_arr = np.asarray(img)

        bbox_image = OffsetImage(img_arr, zoom=1.0)
        bbox_image.image.axes = ax
        bar_length = ax.get_xlim()[1]
        ax.add_artist(AnnotationBbox(bbox_image, (bar_length, name), frameon=False))

    ax.set_xlabel('Nombre d\'apparitions')
    ax.set_ylabel('Nom de l\'invocateur')
    ax.set_title('Les 5 invocateurs les plus courants')

    for spine in ['top', 'right', 'left', 'bottom']:
        ax.spines[spine].set_visible(False)
    ax.tick_params(left=False, bottom=False)

    plt.tight_layout()
    plt.savefig('./results/top_5.png')
    plt.show()
```
